# Feature Importance (SHAP)

```python
import hopsworks
import joblib
import os

project = hopsworks.login()
mr = project.get_model_registry()

fs = project.get_feature_store()
fg = fs.get_feature_group(
    name="aqi_features",   # use your exact feature group name
    version=1
)
df = fg.read()
```

```python
models = mr.get_models("aqi_random_forest")
latest_model = max(models, key=lambda m: m.version)
rf_dir = latest_model.download()
rf_model = joblib.load(os.path.join(rf_dir, "model.pkl"))

print("Loaded version:", latest_model.version)

```

```python
models = mr.get_models("aqi_xgboost")
latest_model = max(models, key=lambda m: m.version)
xgb_dir = latest_model.download()
xgb_model = joblib.load(os.path.join(xgb_dir, "model.pkl"))

print("Loaded version:", latest_model.version)
```

```python
models = mr.get_models("aqi_logistic_regression")
latest_model = max(models, key=lambda m: m.version)
lr_dir = latest_model.download()
lr_model = joblib.load(os.path.join(lr_dir, "model.pkl"))
scaler = joblib.load(os.path.join(lr_dir, "scaler.pkl"))

print("Loaded version:", latest_model.version)
```

```python
features = [
    "pm25","pm10","co","no2","o3","so2","nh3",
    "hour","day","month","day_of_week","is_weekend",
    "aqi_lag_1","aqi_lag_2","aqi_lag_3",
    "aqi_roll_6h","aqi_change_rate",
    "pm25_roll_3h","pm10_roll_6h"
]

X = df[features]

```

```python
import shap

explainer_rf = shap.TreeExplainer(rf_model)
shap_values_rf = explainer_rf.shap_values(X)

shap.summary_plot(shap_values_rf, X, plot_type="bar")

```

```python
explainer_xgb = shap.TreeExplainer(xgb_model)
shap_values_xgb = explainer_xgb.shap_values(X)

shap.summary_plot(shap_values_xgb, X, plot_type="bar")
```
